```python
x = df.iloc[:,2:31].values
y = df.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25,random_state = 0 )
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_train
```

```python
#create a fuction for model
def model(x_train,y_train):
    from sklearn.linear_model import LogisticRegression
    log = LogisticRegression()
    log.fit(x_train,y_train)
     
    from sklearn.tree import DecisionTreeClassifier
    tree = DecisionTreeClassifier()
    tree.fit(x_train,y_train)
    
    from sklearn.ensemble import RandomForestClassifier
    forest = RandomForestClassifier()
    forest.fit(x_train,y_train)
    
    print("[0] Logistic Regression Training Accuracy: ", log.score(x_train,y_train))
    print("[1] Decision Tree Training Accuracy: ", tree.score(x_train,y_train))
    print("[2] Random forest Training Accuracy: ", forest.score(x_train,y_train))
    
    return log, tree, forest
```

```python
models = model(x_train,y_train)
```

```python
#test data on feature matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,models[0].predict(x_test))
```

```python
print(cm)
```

```python
for i in range(len(models)):
        print("model", i)
        cm = confusion_matrix(y_test,models[i].predict(x_test))
        tp = cm[0][0]
        tn = cm[1][1]
        fn = cm[1][0]
        fp = cm[0][1]
        print(cm)
        print("testing accuracy:", (tp+tn)/(tp+tn+fp+fn))
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
for i in range(len(models)):
    print("model no.: ",i)
    print(classification_report(y_test, models[i].predict(x_test)))
    print("accuracy = ", accuracy_score(y_test,models[i].predict(x_test)))
    print("~"*60)
```
